# Image Enhancement and Contrast Enhancement using Gamma Correction and Histogram Equalization

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Load the input image
img = plt.imread('CHNCXR_0004_0.png')

# Perform image enhancement using gamma correction
gamma = 1.5
img_enhanced = np.power(img, gamma)

# Perform contrast enhancement using histogram equalization
img_gray = np.dot(img[...,:3], [0.299, 0.587, 0.114])
img_equalized = np.uint8(255 * (img_gray - img_gray.min()) / (img_gray.max() - img_gray.min()))

# Create histogram plots of the original, enhanced, and equalized images
fig, ax = plt.subplots(3, 2, figsize=(12, 12))
ax[0, 0].imshow(img)
ax[0, 0].set_title('Original Image')
ax[1, 0].imshow(img_enhanced)
ax[1, 0].set_title('Enhanced Image')
ax[2, 0].imshow(img_equalized, cmap='gray')
ax[2, 0].set_title('Equalized Image')

ax[0, 1].hist(img.flatten(), bins=256, range=(0,1))
ax[0, 1].set_title('Histogram of Original Image')
ax[1, 1].hist(img_enhanced.flatten(), bins=256, range=(0,1))
ax[1, 1].set_title('Histogram of Enhanced Image')
ax[2, 1].hist(img_equalized.flatten(), bins=256, range=(0,1))
ax[2, 1].set_title('Histogram of Equalized Image')

# Save the plot to your directory
plt.savefig('my_plot.png')
```

Import the necessary libraries, including numpy and matplotlib, and include the magic command %matplotlib inline to display the plots inline in the notebook.

```python
img = plt.imread('CHNCXR_0004_0.png')
```

Perform image enhancement using gamma correction. Set the value of gamma as per your requirements.

```python
gamma = 1.5
img_enhanced = np.power(img, gamma)
```

Perform contrast enhancement using histogram equalization. Convert the image to grayscale and then use np.uint8() function to convert the data type of the image to 8-bit unsigned integer.

```python
img_gray = np.dot(img[...,:3], [0.299, 0.587, 0.114])
img_equalized = np.uint8(255 * (img_gray - img_gray.min()) / (img_gray.max() - img_gray.min()))
```

To convert the RGB image to grayscale, we can use a weighted average of the three color channels, with weights of 0.299, 0.587, and 0.114 for the red, green, and blue channels respectively. This weighted average formula gives more weight to the green channel, as the human eye is more sensitive to variations in green than in other colors.

The line of code img_gray = np.dot(img[...,:3], [0.299, 0.587, 0.114]) applies this weighted average formula to each pixel of the RGB image, resulting in a grayscale image where each pixel is represented by a single brightness value. The resulting grayscale image is then used for histogram equalization in the subsequent lines of code.

Create a subplot with 3 rows and 2 columns to display the original, enhanced, and equalized images, as well as their respective histograms. Set the size of the figure as 12x12 inches.

Plot the original image in the first row, first column of the subplot. Set the title of the plot as 'Original Image'.

Plot the enhanced image in the second row, first column of the subplot. Set the title of the plot as 'Enhanced Image'.

Plot the equalized image in the third row, first column of the subplot. Set the title of the plot as 'Equalized Image'. Use the 'gray' colormap to display the image in grayscale

```python
ax[2, 0].imshow(img_equalized, cmap='gray')
ax[2, 0].set_title('Equalized Image')
```

Plot the histograms of the original, enhanced, and equalized images in the first, second, and third row of the second column of the subplot, respectively. Set the title of each histogram plot accordingly

```python
ax[0, 1].hist(img.flatten(), bins=256, range=(0,1))
ax[0, 1].set_title('Histogram of Original Image')

ax[1, 1].hist(img_enhanced.flatten(), bins=256, range=(0,1))
ax[1, 1].set_title('Histogram of Enhanced Image')

ax[2, 1].hist(img_equalized.flatten(), bins=256, range=(0,1))
ax[2, 1].set_title('Histogram of Equalized Image')
```
